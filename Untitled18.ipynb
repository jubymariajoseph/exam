```python
df=pd.read_excel("/content/eurovision_1998 to 2012.xlsx")
df
```

```python
df['energy'] = df['energy'].fillna(df['energy'].median())
```

```python
df['duration'] = df['duration'].fillna(df['duration'].median())
df['acousticness'] = df['acousticness'].fillna(df['acousticness'].median())
df['danceability'] = df['danceability'].fillna(df['danceability'].median())
df['tempo'] = df['tempo'].fillna(df['tempo'].median())
df['speechiness'] = df['speechiness'].fillna(df['speechiness'].median())
df['key'] = df['key'].fillna(df['key'].median())
df['liveness'] = df['liveness'].fillna(df['liveness'].median())
df['time_signature'] = df['time_signature'].fillna(df['time_signature'].median())
df['mode'] = df['mode'].fillna(df['mode'].median())
df['loudness'] = df['loudness'].fillna(df['loudness'].median())
df['valence'] = df['valence'].fillna(df['valence'].median())
df['Happiness'] = df['Happiness'].fillna(df['Happiness'].median())
df['Semi.Final.Number'] = df['Semi.Final.Number'].fillna(df['Semi.Final.Number'].mode()[0])
df['Artist.gender'] = df['Artist.gender'].fillna(df['Artist.gender'].mode()[0])
df['Group.Solo'] = df['Group.Solo'].fillna(df['Group.Solo'].mode()[0])
```

Column Semi.Final.Number has 1 or 2 as values so can be filled using mode. Also other than Artist.gender and Group.Solo the rest are numerical and skewed therefore median is used. Group.Solo and Artist.gender are categorical therefore mode is used.

```python
X = df[['Place','Song.Quality','energy','danceability','Semi.Final.Number',
'Song.In.English','duration']]
Y = df[['Points']]
```

```python

X.hist(figsize=(10,10))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
min =MinMaxScaler()
```

```python
x_train[['Place','Song.Quality','energy','danceability','Semi.Final.Number',
'Song.In.English','duration']] = min.fit_transform(x_train[['Place','Song.Quality','energy','danceability','Semi.Final.Number',
'Song.In.English','duration']])
x_test[['Place','Song.Quality','energy','danceability','Semi.Final.Number',
'Song.In.English','duration']] = min.transform(x_test[['Place','Song.Quality','energy','danceability','Semi.Final.Number',
'Song.In.English','duration']])
```

```python
x_test.columns = x_test.columns.str.replace(r'\.+', '_', regex=True)
x_train.columns = x_train.columns.str.replace(r'\.+', '_', regex=True)
```

```python
x_train.columns
```

```python
gr1 = GradientBoostingRegressor(n_estimators=50,learning_rate=0.1,max_depth=4,random_state=42)
gr1.fit(x_train,y_train)
pred1 = gr1.predict(x_test)
MAE1 = mean_absolute_error(y_test,pred1)
print(MAE1)
```

Mean absolute error is 17.32 means the difference between the predicted value and actual value after taking absolte is 17.32
